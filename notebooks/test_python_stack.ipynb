```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = make_classification(n_samples=200, n_features=10, random_state=SEED)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)

model = xgb.XGBClassifier(n_estimators=20, random_state=SEED, verbosity=0, eval_metric='logloss')
model.fit(X_train, y_train)
acc = accuracy_score(y_test, model.predict(X_test))
print(f"XGBoost smoke test — accuracy on synthetic data: {acc:.3f}")
assert acc > 0.5, "XGBoost accuracy below 0.5 on synthetic data — something is wrong"
print("PASS: Python ML stack operational")
```
